The NN



```python
#we are building a not gate (probably the simplest perceptron there is), where for each input it maps to the relavant output (0 -> 1, 1 -> 0)
input_data = [1,0]
output_data = [0,1]
```

```python
#generating random number for weights and biases
import random

weight = random.uniform(0, 1)
bias = random.uniform(0, 1)


print(weight, bias)
```

```python
#defining a simple linear function for the perceptron
def linear(x):
    return ((weight*x)+bias)
```

```python
#defining our ReLu function
def relu(x):
    return max(0,x)
```

```python
#defining our output after relu
def after_relu(x):
    return(relu(linear(x)))
```

Backprop

```python
#Now we define a function to calculate the loss
#here we are basically computing how far is the predicted value (after_relu) from the original value (y)
def loss(x,y):
    return (y-after_relu(x))**2
```

```python
#now to calculate the gradient decent we need to minimize the loss function by updating the weights and biases
#here as we only have one weight, one bias, and one activation function

# d(Loss)/d(actvation)
def drelu(x):
  if (x >= 0):
    return 1


# d(Loss)/d(weight)
def dldw(x, y):
    return 2*(after_relu(x) - y) * drelu(x) * x


# d(Loss)/d(bias)
def dldb(x, y):
    return 2*(after_relu(x) - y) * drelu(x)
```

Train

```python
#defining our learning rate alpha
alpha = 0.01

#arbitrarily choosing 1000 iterations of data
for i in range(1000):
    x = input_data[i%2]
    y = output_data[i%2]
    print("I,O")
    print(x,y)
    print ("--------")
    print ("loss")
    print(loss(x, y))
    print ("--------")
    weight -= alpha * dldw(x, y)
    bias -= alpha * dldb(x, y)



```

```python
#making a predict function (same as after relu, but more accurate as it has been trainer):
def predict(x):
    return(relu(linear(x)))
```
